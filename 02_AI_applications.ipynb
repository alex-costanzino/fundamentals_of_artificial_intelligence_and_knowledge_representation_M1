# **Artificial Intelligence applications**
## Knowledge based systems
Systems whose knowledge is created by humans (domain experts), that cannot be substituted, but supported in long and repetitive tasks.
> "...power of an intelligent system primarily derives by the quantity and quality of knowledge..."

Knowledge changes in time and not alyways is explicit (discovery, mining).

**Decision support systems**: systems that support human experts in their decision making process without replacing it. There are different degree of support:
* **Data analysis for description**: data are used to describe the system, human intervention is largely needed (i.e. configuration under costraints);
* **Diagnosis**: uses data to understand causes, so it performs a *root cause analysis* (i.e. fault diagnosis);
* **Prediction**: uses data to predict future system evolutions, it performs a *what if analysis* (i.e. machine learning classification and regression);
* **Optimization**: to reach decision we need to select the preferred scenario, concept of objective function to minimize or maximize;
* (Decision automation): a step forward that excludes human intervention.

## Formal systems and games
**Minmax algorithm**: designed to define the optimal strategy for one player (max) and for suggesting the best move, hypotesizing that the other player (min) plays at his best. Huge problem space dimension (combinatorial explosion).

Example: [AlphaGO](https://deepmind.com/research/case-studies/alphago-the-story-so-far)

Note that the used [DNN](https://www.youtube.com/watch?v=WSW-5m8lRMs) has learning but not deep reasoning!
